# Data Visualization

### import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### create data frame

```python
url='https://github.com/babayChris/physics77Capstone/raw/refs/heads/main/data/6%20class%20csv.csv'
df=pd.read_csv(url)
```

```python
df.head()
```

```python
temp='Temperature (K)'
L='Luminosity(L/Lo)'
r='Radius(R/Ro)'
abs_mag='Absolute magnitude(Mv)'
st='Star type'
sc='Star color'
spec='Spectral Class'
```

### H-R Diagram

```python
sns.scatterplot(data=df,x=temp,y=L,hue=st,palette='Spectral')
sns.set(rc={'axes.facecolor':'black'})
plt.yscale('log')
plt.gca().invert_xaxis()
plt.title('H-R Diagram')
legend = plt.legend()
frame = legend.get_frame()
frame.set_facecolor('white')
plt.show()
```

<br>

### I. correlation map

```python
sns.pairplot(df.iloc[:,:-3].corr(),height=2, aspect=1.2)#without star type, star color, and spectral class
plt.show()
```

```python
sns.heatmap(df.iloc[:,:-3].corr(),annot=True)
plt.show()
```

### II. scatter/box plots

```python
plt.figure(figsize=(18,8))
plt.subplot(1,3,1)
sns.scatterplot(data=df,x=temp,y=L,hue=st)
plt.subplot(1,3,2)
sns.scatterplot(data=df,x=temp,y=r,hue=st)
plt.subplot(1,3,3)
sns.scatterplot(data=df,x=temp,y=abs_mag,hue=st)
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,3,1)
sns.boxplot(data=df, x=st, y=abs_mag)
plt.subplot(1,3,2)
sns.boxplot(data=df, x=st, y=temp)
plt.yscale('log')
plt.subplot(1,3,3)
sns.boxplot(data=df, x=st, y=L)
plt.yscale('log')

plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,3,1)
sns.boxplot(data=df, x=spec, y=abs_mag)
plt.subplot(1,3,2)
sns.boxplot(data=df, x=spec, y=temp) ## particularly useful, others don't show much
plt.yscale('log')
plt.subplot(1,3,3)
sns.boxplot(data=df, x=spec, y=L)
plt.yscale('log')
```
